# Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
sns.set()
print(f'{matplotlib.__version__}')
```

```python
fig = plt.figure()
fig.show()
```

```python
plt.plot([0,1],[0,3], label='line', color='green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
```

```python
plt.plot([0,1,2],[0,3,2], label='line', color='green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

```python
#fig = plt.figure(dpi = 200)  #figsize=[12,12],
plt.subplot(131)    #1 = liczba wierszy, 2 = liczba kolumn, 1 = konkretny wykres
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,2], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,1,2], 'ro')
```

```python
plt.bar(x=[0,1,2,3], height=[1.2,1.9,2.4,5.1])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2,1.9,2.4,5.1])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2,1.9,2.4,5.1])
```

```python
x1 = np.random.randn(300)
x2= np.random.randn(300)
plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300)*50

plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```

```python
fig = plt.figure(figsize=(8,8))

plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```python

fig = plt.figure(figsize=(12,10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140 + idx+1)
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
```
